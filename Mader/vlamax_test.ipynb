# vlamax.py Test
## importing the file

```python
from vlamax import Model
import numpy as np
```

## create the Model
### create some Data

```python
# Input based on diagnostic
VO2max = 58.6       # ml/kg/min
vLamax = 0.5        # mmol/l/s
RunningEco = 11.95  # ml/kg/min/m/s

# Generate oxygen steady-states
VO2ss = np.arange(1,VO2max-5,0.01) # substract 5 for better visualisation

# Define constants
Ks1 = 0.25 ** 2     # Range: 0.2 - 0.3
Ks2 = 1.2 ** 3      # Range: 1.0 - 1.3
VolRel = 0.45       # Range: 0.40 - 0.45, depends on individual body composition
```

### create the Model

```python
Test = Model(VO2max=VO2max, vLamax=vLamax, RunningEco=RunningEco, Ks1=Ks1, Ks2=Ks2, VolRel=VolRel)
```

### get the calculated attributes

```python
print(f"ADP: {Test.ADP}\nvLass: {Test.vLass}\nLaComb: {Test.LaComb}\nvLanet: {Test.vLanet}")
```

## Functions
### update() and plot_lactate()

```python
Test.update(VO2max = 58.6,       
    vLamax = 0.5,      
    RunningEco = 11.95,
    Ks1 = 0.25,     
    Ks2 = 1.2,    
    VolRel = 0.45)
Test.plot_lactate()
```

### recalculate() and plot_lactate()


```python
Test.vLamax = 0.6
Test.recalculate()
Test.plot_lactate()
```
